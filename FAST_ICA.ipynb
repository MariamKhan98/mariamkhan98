```python
from sklearn.decomposition import FastICA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('attitude.csv')
data.head()
```

```python
data.info()
```

```python
data = data['rownames'].drop
```

```python

data.head()
```

```python
ICA = FastICA(n_components=2)
IndepCompVal = ICA.fit_transform(data.values)
```

```python
print('##Final Indepenedent Components##')
reducedata = pd.DataFrame(data=IndepCompVal, columns=['IC1', 'IC2'])
print(reducedata.head(10))
```
