```python
import numpy as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
```

```python
X, clusters = make_blobs(n_samples=450, centers=6, cluster_std=0.7, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.5, edgecolors='b')
```

```python
#lets implement the birch
brc = Birch(branching_factor=50, n_clusters=None, threshold=1.5)
brc.fit(X)
```

```python
labels = brc.predict(X)
plt.scatter(X[:,0], X[:,1], c=labels, alpha=0.5, edgecolors='b')
```
